```python
import numpy as np
```

```python

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],  # Student 1
    [19, 92, 88],  # Student 2
    [17, 76, 95],  # Student 3
    [18, 65, 70],  # Student 4
    [20, 90, 85]   # Student 5
])
```

```python
# Find the shape of matrix
shape = data.shape
print(shape)

print("Rows : ", shape[0])
print("Columns : ", shape[1])
```

```python
# Find the Average age of students

age_columns = data[0:5,0:1]
print("Age: \n" ,age_columns)
print("\nMean: ", np.mean(age_columns))
```

```python
# Extract Math marks of all students ..

math_marks = data[:,1]
math_marks_2d = data[:,1:2]
print(math_marks)
print(math_marks_2d)
```

```python
# Extract Avg and Std deviation of Science marks 

sci_marks = data[0:6,2]
print(sci_marks)
print("\n")

mean_sci = np.mean(sci_marks)
print("Mean Science: " , mean_sci)

std_dev_sci = np.std(sci_marks)
print("Std Deviation Science: " , std_dev_sci)

var_sci = np.var(sci_marks)
print("Variance Science: " , var_sci)

print("\nMean [Age , Maths , Science] : ", np.mean(data,axis=0))
print("\nStd Dev [Age , Maths , Science] : ", np.std(data,axis=0))
print("\nVariance [Age , Maths , Science] : ", np.var(data,axis=0))
```

```python
# Find the highest Maths Science marks

sci_marks = data[:,2]
maths_marks = data[:,1]

print("Max Marks -> [Maths , Science] " , np.max(sci_marks) , np.max(maths_marks))
```

```python
# Get details of students who scored more than 90 marks in Math

math_marks =data[:,1]
print(math_marks>90)
print(data[math_marks>90])
```

```python
# Increase Math marks of all students by 5

math_marks =data[:,1]
print(data[:,1]+5)
print("\n")
print(data) 

# here data didnt change as we did calculation using data not assigned new values .
# But numpy changing sliced array changes the original array but here we did calculation not actually changes the values 
# changing means assigning new values to an index means maths[1][1] = new_marks will change but doing any calculation
# or operation like data[:,1]+5 will allocate new temp memory for +5 maths marks and 
# delete it immediately after printing it , to assign a new value we must do data[:,1] = data[:,1] + 5

#----------------------

print("\n")
data[:, 1] = data[:, 1] + 5
print(data)

#----------------------

# To prove that slicing of numpy arrays creates a view we will change math_marks[0] to 99
# and check if original data changed 

math_marks[0] = 99
print("\n")
print(data)
```

```python
# Find How many students are younger than 19

bool_age = data[:,0]<19
print(data[bool_age])
print("Count  : " , np.shape(data[bool_age]))
print("Count of students : ", len(data[bool_age]))
```

```python
# Get data of students who scored at least 80 in both subjects

bool_marks_req_data = (data[:,1]>80) & (data[:,2]>80)
print(data[bool_marks_req_data])
```
