```python
import numpy as np
```

```python
#-----------------------GENERATE 1D ARRAY--------------#
arr = np.arange(1,11)
print(arr)
arr1 = np.zeros(6)
print(arr1)
```

```python
#--------------------GENERATE 2D ARRAY OR MATRICES----------------------------#
arr = np.zeros((3,3))
print(arr)
arr1 = np.ones((2,3))
print(arr1)
arr2 = np.zeros((6,10))
print(arr2)
```

```python
#-----------------------LINSPACE FUNCTION---------------------#
arr = np.linspace(1,5,10)
print(arr)
```
