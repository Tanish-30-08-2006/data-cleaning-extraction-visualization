```python
import numpy as np
```

```python
#-----------MATRIX MULTIPLICATION--------------#


arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])
print(f" {arr1} \n\n {arr2}")
```

```python

print("\n Multiplication Based on simple indexing ") 
print(arr1*arr2)


print("\n Multiplication Based on Row-Column method indexing ") 

print(arr1@arr2)
print("\n")
print(np.dot(arr1, arr2))

# np.dot(arr1,arr2) and arr1@arr2 
```

```python
#-------------MATRIX TRANSPOSE---------#

print(arr1.T)
print(arr2.T)
```
